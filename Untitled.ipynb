```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c1 = 0
c2 = 1 / 2
c3 = 2
```

```python
def linear_sin(x0, x1):
    return c1 + c2 * x0 + c3 * np.sin(x1)
```

```python
x_range = np.arange(1, 10, 0.1)
y_range = np.arange(1, 10, 0.1)
xx, yy = np.meshgrid(x_range, y_range)
```

```python
z = linear_sin(xx, yy)
```

```python
fig, ax = plt.subplots(subplot_kw = {'projection': '3d'}, figsize = (12, 8))
ax.plot_surface(xx, yy, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

```python
# Literally the same as the first order taylor expansion, by definition it is!!!

# tangent graph f(x,y) = f(x0,y0) + f_prime_x(x_0, y_0)(x - x0) + f_prime_y(x_0, y_0)(y - y0)
```

```python
class TangentPlane:
    
    def __init__(self, f, f_prime_x, f_prime_y, x_base, y_base):
        
        self.x_base = x_base
        self.y_base = y_base
        
        self.f_base = f(x_base, y_base)
        self.deriv_x_base = f_prime_x(x_base, y_base)
        self.deriv_y_base = f_prime_y(x_base, y_base)
        
    def __call__(self, x, y):
        return self.f_base + self.deriv_x_base * (x - self.x_base) + self.deriv_y_base * (y - self.y_base)
```
